```python
import pandas as pd
```

```python
df= pd.read_csv(r"C:\Users\HDC0422087\Downloads\inventory_dataset - inventory_dataset.csv")
```

```python
df
```

```python
null_columns = df.columns[df.isnull().any()]
```

```python
null_columns
```

```python
median_values = df['Purchase_Price'].median()
df['Purchase_Price'].fillna(median_values, inplace = True)
```

```python
df
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy = 'mean')
```

```python
sale_price_reshaped = df[['Sale_Price']].values
```

```python
df[['Sale_Price']] = imputer.fit_transform(sale_price_reshaped)
```

```python
df
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
Q1 = df['Sale_Price'].quantile(0.25)
Q2 = df['Sale_Price'].quantile(0.75)
IQR = Q2 - Q1
LB = Q1 - 1.5*IQR
UB = Q2 + 1.5*IQR

sales_price = df[(df['Sale_Price'] >= LB) & (df['Sale_Price'] <= UB)]
```

```python
sales_price
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['Sale_Price'] = scaler.fit_transform(df[['Sale_Price']])
```

```python
df
```

```python
df['Sale_Price'] = (df['Sale_Price'] * 100).astype(float)
```

```python
df
```

```python
df
```

```python
df['Discontinued'] = df['Discontinued'].astype(int)
```

```python
df
```

```python
df.to_csv('cleaned_inventory.csv', index=False)
```
