```python
employees_df = payroll_df['Employees']
departments_df = payroll_df['Departments']
```

```python
merged_df = pd.merge(employees_df, departments_df, on='DepartmentID')
merged_df.drop_duplicates(inplace=True)
```

```python
merged_df.head(5)
```

```python
Q1 = merged_df['Salary'].quantile(0.25)
Q3 = merged_df['Salary'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
filtered_df = merged_df[(merged_df['Salary'] > lower_bound) & (merged_df['Salary'] < upper_bound)]
```

```python
filtered_df
```

```python
min_salary = filtered_df['Salary'].min()
max_salary = filtered_df['Salary'].max()
filtered_df['NormalizedSalary'] = (filtered_df['Salary'] - min_salary) / (max_salary - min_salary)
```

```python
filtered_df
```

```python
location_dummies = pd.get_dummies(filtered_df['Location'], prefix='Location')
updated_df = pd.concat([filtered_df, location_dummies], axis=1)
updated_df.drop('Location', axis=1, inplace=True)
```
